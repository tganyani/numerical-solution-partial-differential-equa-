```python
import numpy as np
import matplotlib.pyplot as plt
# import warnings
# warnings.filterwarnings('ignore')
```

```python
f = lambda t,x: np.sqrt(t-x) if x<=t else 0

def F (T,X):
    n = T.shape[0]
    m = T.shape[1]
    u = np.zeros((n,m))
    for i in range(n) :
        for j in range(m):
            u[i,j] = f(T[i,j],X[i,j])
    return  u
```

```python
x = np.linspace(0,12,30)
t = np.linspace(0,8,30)
```

```python
T,X = np.meshgrid(t,x)
U = F(T,X)
```

```python
fig = plt.figure(figsize=(14, 8))
ax = plt.axes(projection='3d')
ax.plot_surface(X, T, U, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')

ax.set_xlabel('X')
ax.set_ylabel('t')
ax.set_zlabel('U')
```

```python
def Zeydeli_Iteration_To_Heat_Equation(n_x,n_t):
    t = np.linspace(0,4,n_t)
    x = np.linspace(0,12,n_x)
    T,X = np.meshgrid(t,x)
    u = F(T,X)
    u[1:,1:]=0
    Itr = 0
    e = 1
    delta_t = t[1]-t[0]
    delta_x = x[1]-x[0]
    #   g
    g = np.zeros((n_x, n_t))
    for i in range(n_x):
        for j in range(n_t):
            if x[i] < t[j]:
                g[i, j] = 3 / (8 * np.sqrt(t[j] - x[i]))
            
    while e>1e-5:
        u2 = np.copy(u)
        for j in range(n_t - 1):
            for i in range(1, n_x - 1):
                k_p = (u[i, j] ** 2 + u[i + 1, j] ** 2) / 4
                k_m = (u[i, j] ** 2 + u[i - 1, j] ** 2) / 4
                laplas = 1 / delta_x ** 2 * (k_p * (u[i + 1, j] - u[i, j]) - k_m * (u[i, j] - u[i - 1, j]))
                u[i, j + 1] = delta_t * laplas + delta_t * g[i, j] + u[i, j]
        Itr+=1
        e = np.linalg.norm(np.ravel(u2-u))
    return T,X,u,
```

```python
R=Zeydeli_Iteration_To_Heat_Equation(30,30)
fig = plt.figure(figsize=(14, 8))
ax = plt.axes(projection='3d')
ax.plot_surface(R[1], R[0], R[2], rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')

ax.set_xlabel('X')
ax.set_ylabel('t')
ax.set_zlabel('U')
```

```python
R=Zeydeli_Iteration_To_Heat_Equation(40,400)
fig = plt.figure(figsize=(14, 8))
ax = plt.axes(projection='3d')
ax.plot_surface(R[1], R[0], R[2], rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')

ax.set_xlabel('X')
ax.set_ylabel('t')
ax.set_zlabel('U')
```

```python
N_x = np.array([10, 20, 30, 40])
N_t = np.array([150, 1000, 2000, 3000])
E = np.zeros(len(N_t))

for k in range(len(N_t)):
    R = Zeydeli_Iteration_To_Heat_Equation(N_x[k],N_t[k])
    e = np.linalg.norm(np.ravel(F(R[1],R[0])-R[2]))/len(np.ravel(R[2]))
    E[k] = e
plt.plot(N_t,E)
plt.xlabel("N_t")
plt.ylabel("E")
plt.legend(["pagreshnost"])
```

```python
def Progonk_To_Heat_Equation(n_x,n_t):
    t = np.linspace(0,4,n_t)
    x = np.linspace(0,12,n_x)
    T,X = np.meshgrid(t,x)
    u = F(T,X)
    u[1:,1:]=0
    Itr = 0
    e = 1
    delta_t = t[1]-t[0]
    delta_x = x[1]-x[0]
    #   g
    g = np.zeros((n_x, n_t))
    for i in range(n_x):
        for j in range(n_t):
            if x[i] < t[j]:
                g[i, j] = 3 / (8 * np.sqrt(t[j] - x[i]))
            
    #     
    D_array = np.zeros((n_x, n_t))
    D_array[:, 0] = - u[:, 0] - g[:, 0] * delta_t
    
    for j in range(n_t - 1):
        
        #     
        A = np.zeros(n_x)
        C = np.zeros(n_x)
        B = np.zeros(n_x)
        D = D_array[:, j]
        
        for i in range(1, len(A) - 1):
            k_p = ((-u[i, j]) ** 2 + (-u[i + 1, j]) ** 2) / 4
            k_m = ((-u[i, j]) ** 2 + (-u[i - 1, j]) ** 2) / 4
            A[i] = k_m * delta_t / (delta_x ** 2)
            B[i] = - (delta_t / delta_x ** 2 * (k_p + k_m) + 1)
            C[i] = k_p * delta_t / (delta_x ** 2)
        
        # 
        N = len(A)
        C_new = [0]
        D_new = [u[0, j]]

        for i in range(1, N - 1):
            current_C = C[i] / (B[i] - A[i] * C_new[i - 1])
            C_new.append(current_C)

        for i in range(1, N - 1):
            current_D = (D[i] - A[i] * D_new[i - 1]) / (B[i] - A[i] * C_new[i - 1])
            D_new.append(current_D)
        
        for i in range(N - 2, 0, -1):
            u[i, j + 1] = D_new[i] - C_new[i] * u[i + 1, j + 1]

        if j < n_t - 1:
            D_array[:, j + 1] = - u[:, j + 1] - g[:, j + 1] * delta_t
    return T,X,u,
```

```python
R=Progonk_To_Heat_Equation(30,30)
fig = plt.figure(figsize=(14, 8))
ax = plt.axes(projection='3d')
ax.plot_surface(R[1], R[0], R[2], rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')

ax.set_xlabel('X')
ax.set_ylabel('t')
ax.set_zlabel('U')
```

```python
R=Progonk_To_Heat_Equation(40,400)
fig = plt.figure(figsize=(14, 8))
ax = plt.axes(projection='3d')
ax.plot_surface(R[1], R[0], R[2], rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')

ax.set_xlabel('X')
ax.set_ylabel('t')
ax.set_zlabel('U')
```

```python
N_x = np.array([10, 20, 30, 40])
N_t = np.array([150, 1000, 2000, 3000])
E = np.zeros(len(N_t))

for k in range(len(N_t)):
    R = Progonk_To_Heat_Equation(N_x[k],N_t[k])
    e = np.linalg.norm(np.ravel(F(R[1],R[0])-R[2]))/len(np.ravel(R[2]))
    E[k] = e
plt.plot(N_t,E)
plt.xlabel("N_t")
plt.ylabel("E")
plt.legend(["pagreshnost-progonk"])
```

```python
def Progonk_Solution_To_Heat_Equation_with_weights(n_x,n_t):
    t = np.linspace(0,4,n_t)
    x = np.linspace(0,12,n_x)
    T,X = np.meshgrid(t,x)
    u = F(T,X)
    u[1:,1:]=0
    delta_t = t[1]-t[0]
    delta_x = x[1]-x[0]
    sigma = 0.5
    #   g
    g = np.zeros((n_x, n_t))
    for i in range(n_x):
        for j in range(n_t):
            if x[i] < t[j]:
                g[i, j] = 3 / (8 * np.sqrt(t[j] - x[i]))


   #     
    D_array = np.zeros((n_x, n_t))
    k_p = (u[0, 0] ** 2 + u[1, 0] ** 2) / 4
    D_array[0, 0] = delta_t * (sigma - 1) / delta_x ** 2 * (k_p * (u[1, 0] - u[0, 0])) - u[0, 0] - delta_t * g[0, 0]
    for i in range(1, n_x - 1):
        k_p = ((u[i, 0]) ** 2 + (u[i + 1, 0]) ** 2) / 4
        k_m = ((u[i, 0]) ** 2 + (u[i - 1, 0]) ** 2) / 4
        D_array[i, 0] = delta_t * (sigma - 1) / delta_x ** 2 * (k_p * (u[i + 1, 0] - u[i, 0]) - k_m * (u[i, 0] - u[i - 1, 0])) - u[i, 0] - delta_t * g[i, 0]
    k_m = (u[-1, 0] ** 2 + u[-2, 0] ** 2) / 4
    D_array[-1, 0] = - delta_t * (sigma - 1) / delta_x ** 2 * k_m * (u[-1, 0] - u[-2, 0]) - u[-1, 0] - delta_t * g[-1, 0]
    
    for j in range(n_x - 1):
        
        #     
        A = np.zeros(n_x)
        C = np.zeros(n_x)
        B = np.zeros(n_x)
        D = D_array[:, j]
        
        for i in range(1, len(A) - 1):
            k_p = ((-u[i, j]) ** 2 + (-u[i + 1, j]) ** 2) / 4
            k_m = ((-u[i, j]) ** 2 + (-u[i - 1, j]) ** 2) / 4
            A[i] = sigma * k_m * delta_t / (delta_x ** 2)
            B[i] = - (sigma * delta_t / delta_x ** 2 * (k_p + k_m) + 1)
            C[i] = sigma * k_p * delta_t / (delta_x ** 2)
        
        # 
        N = len(A)
        C_new = [0]
        D_new = [u[0, j]]

        for i in range(1, N - 1):
            current_C = C[i] / (B[i] - A[i] * C_new[i - 1])
            C_new.append(current_C)

        for i in range(1, N - 1):
            current_D = (D[i] - A[i] * D_new[i - 1]) / (B[i] - A[i] * C_new[i - 1])
            D_new.append(current_D)
        
        for i in range(N - 2, 0, -1):
            u[i, j + 1] = D_new[i] - C_new[i] * u[i + 1, j + 1]
        
        if j < n_t - 1:
            k_p = (u[0, j + 1] ** 2 + u[1, j + 1] ** 2) / 4
            D_array[0, j + 1] = delta_t * (sigma - 1) / delta_x ** 2 * (k_p * (u[1, j + 1] - u[0, j + 1])) - u[0, j + 1] - delta_t * g[0, j + 1]
            for i in range(1, n_x - 1):
                k_p = ((u[i, j + 1]) ** 2 + (u[i + 1, j + 1]) ** 2) / 4
                k_m = ((u[i, j + 1]) ** 2 + (u[i - 1, j + 1]) ** 2) / 4
                D_array[i, j + 1] = delta_t * (sigma - 1) / delta_x ** 2 * (k_p * (u[i + 1, j + 1] - u[i, j + 1]) - k_m * (u[i, j + 1] - u[i - 1, j + 1])) - u[i, j + 1] - delta_t * g[i, j + 1]
            k_m = (u[-1, j + 1] ** 2 + u[-2, j + 1] ** 2) / 4
            D_array[-1, j + 1] = - delta_t * (sigma - 1) / delta_x ** 2 * k_m * (u[-1, j + 1] - u[-2, j + 1]) - u[-1, j + 1] - delta_t * g[-1, j + 1]
    return T,X,u
```

```python
R=Progonk_Solution_To_Heat_Equation_with_weights(30,30)
fig = plt.figure(figsize=(14, 8))
ax = plt.axes(projection='3d')
ax.plot_surface(R[1], R[0], R[2], rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')

ax.set_xlabel('X')
ax.set_ylabel('t')
ax.set_zlabel('U')
```

```python
N_x = np.array([10, 20, 30, 40])
N_t = np.array([150, 1000, 2000, 3000])
E = np.zeros(len(N_t))

for k in range(len(N_t)):
    R = Progonk_To_Heat_Equation(N_x[k],N_t[k])
    e = np.linalg.norm(np.ravel(F(R[1],R[0])-R[2]))/len(np.ravel(R[2]))
    E[k] = e
plt.plot(N_t,E)
plt.xlabel("N_t")
plt.ylabel("E")
plt.legend(["pagreshnost-progonk-with weights"])
```
