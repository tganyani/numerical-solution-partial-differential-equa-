```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def Yakob(A,b):
    dim = len(A)
    D = np.zeros((dim,dim))
    T=  np.array(A)
    sol = np.array(np.zeros(dim).reshape(dim,1))
    E = [1]
    for i in range(dim):
        T[i,i] = 0
        D[i,i] = A[i,i]
    for i in range(dim):
        if D[i,i] == 0:
            return f"Диагональные элементы не должны быть равны нулю, попробуйте поменять местами строку по индексу {i+1}."
        if np.sum(np.abs(T[i,:])) > np.abs(D[i,i]):
            return  " Итерация может не сходится",i
    while  E[-1]> 1e-4:
        X_k = np.matmul(np.linalg.inv(D),b.reshape(dim,1)-np.matmul(T,sol[:,-1].reshape(dim,1)))
        sol = np.hstack((sol,X_k))
        e = np.sqrt(np.sum(np.square(sol[:,-1]-sol[:,-2])))
        E.append(e)
    return sol[:,-1]
```

```python
def Numerical_Solution_Laplace_Equation(n):
    A = np.zeros((n,n)) 
    n_ = int(np.sqrt(n))
    for i in range(n):
        A[i,i] = 4
        if(i!=n-1):
            if((i+1)%(np.sqrt(n)) != 0):
                A[i,i+1]=-1
            else:
                A[i,i+1]=0
        if(i < n-np.sqrt(n)):
            A[i,i+n_] = -1
    B=A.T
    C = A + B
    for i in range(n):
        C[i,i] = 4
    u_0y= lambda y:-y**2   #a
    u_1y = lambda y: 1-y**2  #d
    u_x0 = lambda x:x**2  #b
    u_x1 = lambda x:x**2-1  #c
    x = np.linspace(0,1,n_+2)
    y = np.linspace(0,1,n_+2)
    X,Y=np.meshgrid(x,y)
    YY= Y[::-1]
    b = np.zeros(n)
    b[0] = u_0y(YY[1,0]) + u_x1(X[0,1]) #'a+c'
    b[n_-1] =u_x1(X[0,n_]) + u_1y(YY[1,0])#'c+d'
    b[n-n_] =u_0y(YY[n_,0]) + u_x0(X[0,1]) #'a+b'
    b[n-1] = u_x0(X[0,n_]) + u_1y(YY[n_,0]) #'b+d'
    for i in range(n):
        if ( i>0 and i<n_ -1):
            b[i]= u_x1(X[1,i+1]) #c
        if(i>n_-1 and i <n-n_ ):
            if(i%n_==0):
                b[i]=u_0y(YY[int(i/n_)+1,1]) #a
            elif(i%n_ == n_-1):
                b[i]= u_1y(YY[int(i//n_)+1,1]) #d
            else:
                b[i]=0
        if ( i>n-n_ and i<n -1):
            b[i]= u_x0(X[0,int(i%n_)+1]) #b
            
    sol =Yakob(C,b) #np.linalg.solve(C,b)
    W= np.zeros((n_+2,n_+2))
    W[1:-1,1:-1] = sol.reshape(n_,n_)
    W[0,:]=u_x1(X[0,:])
    W[:,0]= u_0y(YY[:,0])
    W[:,-1]= u_1y(YY[:,0])
    W[-1,:]=u_x0(X[0,:])
    return X,YY,W
```

```python
# при n = 3
R=Numerical_Solution_Laplace_Equation(9)
fig = plt.figure(figsize=(14, 8))
ax = plt.axes(projection='3d')
ax.plot_surface(R[0], R[1], R[2], rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_title('$n=3$');
```

```python
f_xy = lambda x,y:x**2-y**2

def Error(N):
    n= len(N)
    E = np.zeros(n)
    for i in range(n):
        R=Numerical_Solution_Laplace_Equation(N[i])
        WW=f_xy(R[0],R[1])
        e=np.linalg.norm(np.ravel(WW-R[2]))/(N[i]*np.linalg.norm(np.ravel(WW)))
        E[i]=e
    return E
```

```python
N =np.array([36,49,64,81,100,144,169])
E=Error(N)
E
```

```python
plt.plot(N,E)
plt.xlabel('$N $')
plt.ylabel('$E $')
```
