```python

f_xy = lambda x,y: x**2 - y**2

x = np.linspace(0,1,50)
y = np.linspace(0,1,50)
X, Y = np.meshgrid(x, y)
Z = f_xy(X,Y)
```

# Задание 2

решение слау методом Якоби

$$
Ax=b
$$

превращая матрицу $A$ в сумму двух матриц $D+T$

где $D=\begin{pmatrix}
a_{11} & 0 & \ldots & 0\\
\vdots & a_{11}& \ddots & 0 \\
0 & \ldots& \ldots & a_{nn}
\end{pmatrix}$

,$T=\begin{pmatrix}
0 & a_{12} & \ldots & a_{1n}\\
\vdots & 0& \ddots & a_{2n} \\
a_{n1} & \ldots& \ldots &0
\end{pmatrix}$

$
(D+T)x=b \space \space\space\space ,Dx = b-Tx  \space \space\space\space ,x = D^{-1}(b-Tx)
$

$ x_{k+1} = D^{-1}(b-Tx_{k}) $

$
a_{ii} \neq 0 \space \space \space i=\overline {1,n}
$

$
\sum_{j \neq i}^{n}| a_{i,j}| < {| a_{ii} |} \space \space \space i=\overline {1,n}
$

```python
A = np.array([
    [-23,-7,5,2],
    [-7,-21,4,9],
    [9,5,-31,-8],
    [0,1,-2,10]
])
```

```python
A
```

```python
b =np.array([-26,-55,-58,-24])
```

```python
def Yakob(A,b):
    dim = len(A)
    D = np.zeros((dim,dim))
    T=  np.array(A)
    sol = np.array(np.zeros(dim).reshape(dim,1))
    E = [1]
    for i in range(dim):
        T[i,i] = 0
        D[i,i] = A[i,i]
    for i in range(dim):
        if D[i,i] == 0:
            return f"Диагональные элементы не должны быть равны нулю, попробуйте поменять местами строку по индексу {i+1}."
        if np.sum(np.abs(T[i,:])) > np.abs(D[i,i]):
            return  " Итерация может не сходится",i
    n = 0
    while  E[-1]> 1e-4:
        X_k = np.matmul(np.linalg.inv(D),b.reshape(dim,1)-np.matmul(T,sol[:,-1].reshape(dim,1)))
        sol = np.hstack((sol,X_k))
        e = np.sqrt(np.sum(np.square(sol[:,-1]-sol[:,-2])))
        E.append(e)
        n+=1
    return sol[:,-1], n,E
                        
        
```

```python
Yakob(A,b)
```
